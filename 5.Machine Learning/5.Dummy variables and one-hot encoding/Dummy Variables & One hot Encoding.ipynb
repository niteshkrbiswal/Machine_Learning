multicolinearity - (where 2 or more dummy variables directly relational to each other)

To Avoid multicolinearity we have to drop one dummy variable ,so that thier sum wont be always same every time

## 4. OneHotEncoding with Top Categories
###### When we have lot of categories in catgeorical column we try to club some low volumn categories for easy calculation

```python
counts = df['brand'].value_counts()
counts
```

```python
df['brand'].nunique()
```

```python
threshold = 100
# It means before 100 all column will be individual but after 100 it will group as 1 
```

```python
repl = counts[counts <= threshold].index
repl

#Car whose threshold is less than 100
#which we have to replace with a single column
```

```python
pd.get_dummies(df['brand'].replace(repl, 'other_cars')).sample(5)
```
